```python
# Using the data from the insurance.csv dataset to determine relation between age and insurance charges

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Using pandas to read the insurance.csv file and storing the file as variable

df = pd.read_csv("insurance.csv")
df.head()
```

```python
# Using the iloc function to locate the age and charges column fron the dataframe

x = df.iloc[:,0].values  # locates all values from the age column
y = df.iloc[:,-1].values   # locates all values from the charges column

# Using above located values to produce a scatter plot with Age on x axis and Insurance charges on y axis

plt.scatter(x,y,color = 'g')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.show()
```

```python
# Using sklearn to fit linear regression model and make predictions on the data

model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)
```

```python
# Predicting the charges for a person of age 50

unknown_age = [[50]]

x_pred = np.append(x , unknown_age).reshape(-1,1)
y_pred = model.predict(x_pred)

print("Insurance charge for a person of age 50 should be:" , model.predict(unknown_age))
```

```python
# Scatter plot with best fit line

plt.scatter(x,y,color = 'g')
plt.plot(x_pred,y_pred,color = 'b')
plt.xlabel('Age')
plt.ylabel('Insurance Charge')
plt.show()
```
