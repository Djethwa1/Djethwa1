```python
# importing the spacy library and loading the small packeage on the NLP english language model

import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# Removing stop words, null values, punctuations in the dataset

def preprocess(text):
    doc = nlp(text)
    return''.join([token.lemma_lower().strip().str for token in doc if not token.is_stop and not token.is_punct])
```
